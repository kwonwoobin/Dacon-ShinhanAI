# 스크래피 가온데이터 합치기
****week_num datatype int로 통일하고 합치기

```python
import pandas as pd
import os
```

```python
dir_name = '2022'
gaon_data_list = os.listdir(f'./gaon_data/{dir_name}')
gaon_data_list
```

```python

list_tmp_df = []
for data in gaon_data_list:
    df = pd.read_csv(f'./gaon_data/{dir_name}/{data}', index_col=0)
    df = df[:100]

    # 글자 행 삭제
    drop_index = df.loc[df['year']=='year'].index
    df = df.drop(drop_index, axis=0)

    list_tmp_df.append(df)
    
gaon_df = pd.concat(list_tmp_df, ignore_index=True)
gaon_df = gaon_df[['year', 'week_num', 'start_date', 'end_date', 'rank', 'music', 'singer', 'album']]
gaon_df.to_csv(f'./gaon_data/{dir_name}.csv')
```

```python
for dir_name in [2021,2020,2019,2018,2017,2016,2015,2014,2013]:
    gaon_data_list = os.listdir(f'./gaon_data/{dir_name}')
    gaon_data_list


    list_tmp_df = []
    for data in gaon_data_list:
        df = pd.read_csv(f'./gaon_data/{dir_name}/{data}', index_col=0)
        df = df[:100]

        try:
            # 글자 행 삭제
            drop_index = df.loc[df['year']=='year'].index
            df = df.drop(drop_index, axis=0)
        except:
            pass

        list_tmp_df.append(df)
        
    gaon_df = pd.concat(list_tmp_df, ignore_index=True)
    gaon_df = gaon_df[['year', 'week_num', 'start_date', 'end_date', 'rank', 'music', 'singer', 'album']]
    gaon_df.to_csv(f'./gaon_data/{dir_name}.csv')
```

# 스크래피로 모은 데이터 concat code

```python


list_tmp_df = []
for data in gaon_data_list:
    df = pd.read_csv(f'./gaon_data/{data}')

    # week_num이라는 값이 들어가있는 행 삭제 후 week_num값을 int로 형변환
    drop_index = df.loc[df['week_num']=='week_num'].index
    df = df.drop(drop_index, axis=0)

    # 1.week_num 형변환
    df = df.astype({'week_num':'int32', 'year':'int32'})

    data = data.rstrip('.csv')
    data = data.split('_')
    if len(data) == 3:
        start_week = int(data[2])
        end_week = int(data[2])
    else:
        start_week, end_week = int(data[2]), int(data[3])
        
    df = df.loc[(df['week_num'] >= start_week) & (df['week_num'] <= end_week)]
    list_tmp_df.append(df)
    
gaon_df = pd.concat(list_tmp_df, ignore_index=True)
gaon_df
```
