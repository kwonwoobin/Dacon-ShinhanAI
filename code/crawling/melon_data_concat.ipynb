```python
import pandas as pd
import os
```

```python
melon_data_list = os.listdir('../code/melon_data')
```

```python
melon_data_list
```

```python

list_tmp_df = []
for data in melon_data_list:
    df = pd.read_csv(f'../code/melon_data/{data}')

    # 글자 행 삭제 후 week_num값을 int로 형변환
    drop_index = df.loc[df['music']=='music'].index
    df = df.drop(drop_index, axis=0)

    list_tmp_df.append(df)

melon_df = pd.concat(list_tmp_df, ignore_index=True)
melon_df
```

```python
melon_df = melon_df[['music','singer','music_id','release_date','category','music_like','comment_count']]
```

```python
list_tmp_df = []
for data in melon_data_list[:-1]:
    df = pd.read_csv(f'./melon_data/{data}')

    # week_num이라는 값이 들어가있는 행 삭제 후 week_num값을 int로 형변환
    drop_index = df.loc[df['week_num']=='week_num'].index
    df = df.drop(drop_index, axis=0)

    # 1.week_num 형변환
    df = df.astype({'week_num':'int32', 'year':'int32'})

    data = data.rstrip('.csv')
    data = data.split('_')
    if len(data) == 3:
        start_week = int(data[2])
        end_week = int(data[2])
    else:
        start_week, end_week = int(data[2]), int(data[3])
        
    df = df.loc[(df['week_num'] >= start_week) & (df['week_num'] <= end_week)]
    list_tmp_df.append(df)
    
gaon_df = pd.concat(list_tmp_df, ignore_index=True)
gaon_df
```
